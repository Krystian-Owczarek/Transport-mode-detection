Jest problem z danymi. W tej tabeli mamy tylko i wyłącznie przypadki gdzie wartość minimalna i maksymalna są równe. Więc odchylenie standardowe powinno być dla nich równe 0. A jednak w kolumnie ASRV_STD są wartości odbiegające od zera. Przyczyną może być fakt, że wartości różnią się dopiero po 6 liczbie po przecinku.

Dodatkowo w kolumnach STD z innych sensorów widzimy różnicę między MIN i MAX. Mówi nam to, że problematyczne przypadki pochodzą jedynie z tego sensora, przez co pozbywając się ich, stracilibyśmy wartościowe dane.

Jednak zanim to zrobimy potrzebujemy punktu odniesienia. Wyszkolimy więc model przyjmując założenia z poprzedniego notatnika. Jednakże tym razem, zamiast standaryzacji przeprowadzimy normalizację, ponieważ nasze dane nie układają się wobec rozkładu normalnego.

```python
learning_data = df.copy()
learning_data['time'] = learning_data['time'].astype(int)
learning_data['ASAperMINUTE'] = learning_data['ASA_MEAN']/learning_data['time']
learning_data['ASGperMINUTE'] = learning_data['ASA_MEAN']/learning_data['time']
y = learning_data['target']
X = learning_data.drop(['target', 'SND_MIN', 'SND_MEAN'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# ALGORYTMY
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

# GridSearch i Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import PolynomialFeatures, Normalizer

# Ocena
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, plot_confusion_matrix, plot_roc_curve
from yellowbrick.classifier import ROCAUC
from mlxtend.plotting import plot_decision_regions
```

```python
num_feat = X_train.columns

num_prep = Pipeline(steps=[
    ('polynomial_features', PolynomialFeatures()),
    ('scaler_1', Normalizer())
])

data_prep = ColumnTransformer(transformers=[
    ('numeric_preprocessing', num_prep, num_feat)
])
```

```python
def teach_and_check_scores(X_train, y_train, X_test, y_test, model):
    
    model.fit(X_train, y_train)
    
    preds = model.predict(X_test)
    
    preds_proba = model.predict_proba(X_test)    
    
    print(classification_report(y_test, preds))
    
    print(f'ROC AUC SCORE: {roc_auc_score(y_test, preds_proba, multi_class="ovr")}')

    plot_confusion_matrix(model, X_test, y_test, cmap='Blues')
```

# BAZOWY

```python
knn_clf = Pipeline(steps=[('preprocessor', data_prep),
                          ('model', KNeighborsClassifier())])
```

```python
params = {
    'preprocessor__numeric_preprocessing__polynomial_features__degree': [2, 3],
    'model__n_neighbors': [3, 5, 7],
    'model__n_jobs': [-1, 200, 500, 1000],
    'model__weights': ['uniform', 'distance']
}

grid_search = GridSearchCV(knn_clf, params, cv=5, n_jobs=-1, verbose=10, scoring='f1_macro')
grid_search.fit(X_train, y_train)
print('Wybrane hiperparametry: ', grid_search.best_params_)
knn_v1 = grid_search.best_estimator_
```

```python
teach_and_check_scores(X_train, y_train, X_test, y_test, knn_v1)
```

Model radzi sobie gorzej niż przy użyciu mniejszej ilości danych...

```python
visualizer = ROCAUC(knn_v1, classes=np.unique(y_test))
visualizer.fit(X_train, y_train)
visualizer.score(X_test, y_test)
visualizer.show()
```

Model z nowymi danymi uzyskał o 8% niższy wynik od modelu bazowego z poprzedniego notatnika.
